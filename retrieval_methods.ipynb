```python
import os
from langchain_google_genai import GoogleGenerativeAIEmbeddings
from langchain_community.vectorstores import Chroma
from dotenv import load_dotenv
import google.generativeai as genai

load_dotenv()
api=os.getenv('GOOGLE_API_KEY')
genai.configure(api_key=api)
```

```python
persist_directory='docs/chroma/'
embeddings=GoogleGenerativeAIEmbeddings(model="gemini-embedding-2-preview",
                                        # max_retries=5,timeout=60
)
print('Loading existing vector db from disk..')

vectordb=Chroma(
    persist_directory=persist_directory,
    embedding_function=embeddings
)

docs=vectordb.similarity_search('What are the main findings in this document?',k=3)
for i, doc in enumerate(docs):
    print(f"\n--- Chunk {i+1} ---")
    print(doc.page_content)
```

```python
docs=vectordb.similarity_search('Show me 10 clauses from this document',k=3)
for i, doc in enumerate(docs):
    print(f"\n--- Chunk {i+1} ---")
    print(doc.page_content)
```

```python
docs=vectordb.similarity_search('What is the latest sales tax rate?',k=3)
for i, doc in enumerate(docs):
    print(f"\n--- Chunk {i+1} ---")
    print(doc.page_content)
```

```python
texts = [
    """The Amanita phalloides has a large and imposing epigeous (aboveground) fruiting body (basidiocarp).""",
    """A mushroom with a large fruiting body is the Amanita phalloides. Some varieties are all-white.""",
    """A. phalloides, a.k.a Death Cap, is one of the most poisonous of all known mushrooms.""",
]

smalldb=Chroma.from_texts(texts,embedding=embeddings)
```

```python
smalldb.similarity_search('Tell me about all-white mushrooms with large fruiting bodies',k=2)
```

```python
smalldb.max_marginal_relevance_search('Tell me about all-white mushrooms with large fruiting bodies',k=2,fetch_k=3)
```

### Addressing Diversity: Maximum marginal relevance
 
`Maximum marginal relevance` strives to achieve both relevance to the query *and diversity* among the results.

```python
docs=vectordb.max_marginal_relevance_search('What is the latest sales tax rate?',k=3)
for i, doc in enumerate(docs):
    print(f"\n--- Chunk {i+1} ---")
    print(doc.page_content)
```

```python
from langchain_community.document_loaders import PyPDFLoader

loader=PyPDFLoader('Sales_Tax_1990_Updated.pdf')
pages=loader.load()

all_page_text=[p.page_content for p in pages]
joined_page_text=" ".join(all_page_text)
```

```python
from langchain_text_splitters import RecursiveCharacterTextSplitter

r_splitter=RecursiveCharacterTextSplitter(
    chunk_size=1500,
    chunk_overlap=150
)

splits = r_splitter.split_text(joined_page_text)
```

```python
from langchain_community.retrievers import SVMRetriever, TFIDFRetriever

tfidf=TFIDFRetriever.from_texts(splits)
```

```python
#Reduce batch size and increase wait/sleep time if you run into error: Resource Exhausted
import time

BATCH_SIZE = 10
all_embeddings = []

print(f"Total splits to embed: {len(splits)}")

for i in range(0, len(splits), BATCH_SIZE):
    batch = splits[i : i + BATCH_SIZE]
    
    print(f"Embedding batch {i} to {i + len(batch)}...")
    batch_embeddings = embeddings.embed_documents(batch)
    
    # Add them to our master list
    all_embeddings.extend(batch_embeddings)
    
    time.sleep(8)  

print("Embedding complete! Initializing Retriever...")

# Initialize the SVMRetriever manually
svm = SVMRetriever(
    texts=splits, 
    embeddings=embeddings, 
    index=all_embeddings # We pass our pre-computed embeddings here
)

print("Retriever ready.")
```

```python
question = "What are major topics in this document?"
docs_tfidf=tfidf.invoke(question)
print(docs_tfidf[0])
```

```python
question = "What are major topics in this document?"
docs_svm=svm.invoke(question)
print(docs_svm[0])
```

## Recursive splitting details

`RecursiveCharacterTextSplitter` is recommended for generic text. 

```python
some_text="""Have you ever wondered why we see so few stars in the night sky today?
When our ancestors, or even our parents, were young, the night sky was filled with countless stars,
but today only a handful are visible from our cities. The reason is light pollution. The Intense artificial light from our homes,
streetlights, and buildings brightens the atmosphere, effectively washing out the natural light of the stars."""
```

```python
c_splitter=CharacterTextSplitter(
    chunk_size=450,
    chunk_overlap=0,
    separator = ' '
    # length_function=len
)

r_splitter = RecursiveCharacterTextSplitter(
    chunk_size=450,
    chunk_overlap=0, 
    separators=["\n\n", "\n", " ", ""]
)

c_splitter.split_text(some_text)
```

```python
r_splitter.split_text(some_text)
```

## Context aware splitting
Use `MarkdownHeaderTextSplitter` to preserve header metadata in our chunks

```python
from langchain_text_splitters import MarkdownHeaderTextSplitter

markdown_document = """# Title\n\n \
## Chapter 1\n\n \
Hi this is Jim\n\n Hi this is Joe\n\n \
### Section \n\n \
Hi this is Lance \n\n 
## Chapter 2\n\n \
Hi this is Molly"""

headers_to_split_on = [
    ("#", "Header 1"),
    ("##", "Header 2"),
    ("###", "Header 3"),
]

markdown_splitter = MarkdownHeaderTextSplitter(
    headers_to_split_on=headers_to_split_on
)
md_header_splits = markdown_splitter.split_text(markdown_document)
```

# Embeddings

```python
from langchain_google_genai import GoogleGenerativeAIEmbeddings

embeddings=GoogleGenerativeAIEmbeddings(model="gemini-embedding-2-preview",
                                        # max_retries=5,timeout=60
)
```

```python
import numpy as np

vector1 = embeddings.embed_query("hello, world!")
vector2 = embeddings.embed_query("bye, world!")
vector3 = embeddings.embed_query("goodbye, world!")

np.dot(vector1,vector2)
```

```python
text_splitter = RecursiveCharacterTextSplitter(
        chunk_size = 1500,
        chunk_overlap = 150
    )
   
splits = text_splitter.split_documents(pages)
```

```python
import time
from langchain_community.vectorstores import Chroma

persist_directory = 'docs/chroma/'

# Fix 1: Lower the batch size drastically to prevent hitting the token/RPM ceiling
BATCH_SIZE = 5 
WAIT_TIME = 10  # Since we don't have auto-retries, give the API a comfortable 10s breathing room

# Initialize the vector store with the first batch
if splits:
    print(f"Initializing vector store with the first {min(BATCH_SIZE, len(splits))} documents...")
    vectordb = Chroma.from_documents(
        documents=splits[:BATCH_SIZE],
        embedding=embeddings,
        persist_directory=persist_directory
    )
    
    # Process the remaining documents in controlled small batches
    for i in range(BATCH_SIZE, len(splits), BATCH_SIZE):
        batch = splits[i:i + BATCH_SIZE]
        print(f"Uploading batch {i // BATCH_SIZE + 1} ({len(batch)} documents)...")
        
        # 👇 You nailed it. This line is mandatory here!
        vectordb.add_documents(documents=batch)
        
        print(f"Waiting {WAIT_TIME} seconds to avoid rate limits...")
        time.sleep(WAIT_TIME)
        
    print("All documents successfully embedded and stored!")
```

```python
vectordb._collection.count()
```

```python
question='Is there any email I can ask for help?'

docs=vectordb.similarity_search(question,k=3)
```
